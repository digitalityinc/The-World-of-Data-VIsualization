```python
import pandas as pd
import numpy as np
```

```python
# get current size
fig_size = plt.rcParams["figure.figsize"]
print ("Current size:", fig_size)
# let's make the plots a bit bigger than the default
# set figure width to 14 and height to 6
fig_size[0] = 14
fig_size[1] = 6
plt.rcParams["figure.figsize"] = fig_size
print ("Current size:", fig_size)
```

```python
#numpy arrays or pd.Series.
x = [1, 2, 3, 4, 5]
y = [5000, 9000, 12000, 6000, 4000]
plt.plot(x,y, color='blue', lw=4, alpha=0.5, label='Daily level step count')
plt.axis(xmin=0, xmax=6, ymin=4000, ymax=15000)

plt.xticks(x, ('2023-08-21','2023-08-22', '2023-08-23', '2023-08-24', '2023-08-25'))


plt.grid(True)
plt.legend(loc='upper right', prop={'size':30})
plt.ylabel('step count', size=16, labelpad=20)
plt.xlabel('date', size=16, labelpad=20)
plt.title('Daily step count plotting', size=18)

#plt.annotate(str(12000), xy=(3,12000+5), fontsize=15)
for i,j in zip(x,y):
    inplot = plt.annotate(str(j), xy=(i,j+5), fontsize=15, color='black')
    
    

plt.axvline(x[1], color='red', linestyle='--', linewidth=1)
plt.axhline(y[1], color='orange', linestyle='--', linewidth=1)


```
